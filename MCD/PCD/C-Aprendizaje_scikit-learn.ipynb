```python
import requests
import pandas as pd
```

```python
url='https://datos.sonora.gob.mx/dataset/940de7d8-73ec-4a08-8d26-b7cc97ff17bd/resource/46977217-9396-4f9c-bddb-6616e82530b4/download/202209281437030-file.xlsx'
```

```python
# Download the file
response = requests.get(url)
response.raise_for_status()  # Raise an exception for HTTP errors

# Save the Excel file locally (optional)
file_name = '202209281437030-file.xlsx'
with open(file_name, 'wb') as file:
    file.write(response.content)

# Read the Excel file into a Pandas DataFrame
# Replace 'Sheet1' with the name of the sheet you want to read, if needed
data = pd.read_excel(file_name)#, sheet_name='Sheet1')

# Display the first few rows of the DataFrame
print(data.head())
```
